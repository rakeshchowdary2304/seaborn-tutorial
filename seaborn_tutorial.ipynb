```python
import pandas as pd
import seaborn as sb
from matplotlib import pyplot as plt
df=sb.load_dataset('tips')
print(df.to_string())
```

```python
g=sb.FacetGrid(df,col="time")
g.map(plt.hist,"tip")
plt.show()
```

```python
g=sb.FacetGrid(df,col="day")
g.map(plt.hist,"tip")
plt.show()
```

```python
g=sb.FacetGrid(df,col="sex",hue="smoker")
g.map(plt.scatter,"total_bill","tip")
plt.show()
```

```python
g=sb.FacetGrid(df,col="sex",hue="day")
g.map(plt.scatter,"total_bill","tip")
plt.show()
```

```python
sb.barplot(x="sex",y="tip",hue="time",data=df)
plt.show
```

```python
sb.barplot(x="sex",y="tip",data=df)
plt.show
```

```python
sb.regplot(x="total_bill",y="tip",data=df)
plt.show()
```

```python
sb.pairplot(df)
plt.show()
```

```python
selected_vars=['total_bill','tip']
sb.pairplot(df,vars=selected_vars)
plt.show()
```

```python
df=sb.load_dataset('tips')
sb.pairplot(df,hue='size')
plt.show()
```

```python
df=sb.load_dataset('tips')
sb.pairplot(df,hue='size',palette="husl")
plt.show()
```

```python
df=sb.load_dataset('tips')
sb.pairplot(df,diag_kind='kde')
plt.show()
```

```python
sb.pairplot(df,kind='reg')
plt.show()
```

```python
sb.pairplot(df,kind='kde')
plt.show()
```
